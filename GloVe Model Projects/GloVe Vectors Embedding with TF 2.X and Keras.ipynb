```python
#knowing maximum index of encoded_text having maximum length of array

max_ = []
def get_max_ind(x):
    for i in x[:]:
        max_.append(len(i))
    k=0   
    for j in max_:
        if j == max_length:
            index = k
        k+=1 
    return index
max_index =get_max_ind(encoded_text)
print(max_index)
```

# GloVe Vectors

```python
glove_vectors = dict()
```

Now we have to make a dictionary as `{'words' : vectors of each word}` from `GloVe models/glove.6B.200d.txt` file (200d - 200dimensional)

```python
file = open('GloVe models/glove.6B.200d.txt', encoding = 'utf-8')

for line in file:
    values = line.split()
    word = values[0]
    vectors = np.asarray(values[1:])
    glove_vectors[word] = vectors
file.close()
```

## matching and making a vector of words present in glove_vectors

```python
word_vector_matrix = np.zeros((vocab_size, max_len))
word_vector_matrix
```

```python
for word, index in token.word_index.items():
    vector = glove_vectors.get(word)
    
    if vector is not None:
        word_vector_matrix[index] = vector
    else:
        print(word)
```

word_vector_matrix -- contains the weights from glove vectors for each matched word in X and remaining unmatched words make into zeroes...


```python

word_vector_matrix
```

```python
word_vector_matrix.shape
```

word_vector_matrix --- It has the weights for corresponding numbers in X

## Model Building with Keras

```python
X
```

```python
X_train ,X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size = 0.2, stratify = y)
```

```python
vec_size=200 
#changing each word in an index into a vector of size 200
```

```python
model = Sequential()
model.add(Embedding(vocab_size, vec_size, input_length = max_len, weights = [word_vector_matrix], trainable = False))

model.add(Conv1D(64, 8, activation = 'relu'))
model.add(Conv1D(64, 8, activation = 'relu'))
model.add(MaxPooling1D(2))

model.add(Flatten())
model.add(Dense(32, kernel_initializer = 'he_normal',activation = 'relu'))
model.add(Dropout(0.25))

model.add(Dense(16, kernel_initializer = 'he_normal', activation='relu'))
#model.add(GlobalMaxPooling1D())

model.add(Dense(1, activation='sigmoid'))

```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs = 12, validation_data = (X_test,y_test))
```
